```python
import tkinter as tk
# from tkinter import ttk

todo_list = ["Lunch at 12:00 today", "Meeting preparation",
             "Third Item", "Fourth Item", "Fifth Item",
             "Sixth Item", "Seventh Item", "Eighth Item",
             "Ninth Item", "Tenth Item", "Eleventh Item",
             "Twelveth Item", "Thirteenth Item"] 

todo_list_complete = []

# This list will hold the GUI widgets that hold all the 
# todos on the screen.
todo_list_display = []

# main window
window = tk.Tk()
window.title("TODO App")


todo_string = tk.StringVar()

todo_text = tk.Entry(window, textvariable = todo_string)

todo_text.grid(row=0, column = 0)

def addClicked () :
    todo_list.append(todo_string.get())
    render_list_items(todo_list)

add_button = tk.Button(window, text="Add", command=addClicked)

add_button.grid(row=0, column = 1)


## Add a horizontal line
separator = tk.ttk.Separator(window)

## The width of the line is 1 px and length
separator.grid(row=1,columnspan=2, sticky="ew", pady=10)

def mark_complete_clicked(index):

    # add to todo_list_complete
    todo_list_complete.append(todo_list[index])

    # remove the clicked element from todo_list
    del todo_list[index]

    # re-render the list items
    render_list_items(todo_list)

def render_list_items(todo_list):

    ################## CHANGE START ############################################
    # if there are existing todo list items (widgets), destroy them
    for index, entry in enumerate(todo_list_display):
        # entry[0].set('Hello') # cannot destroy string vars.
        entry[1].destroy()
        entry[2].destroy()

    todo_list_display.clear()
    
    ################## CHANGE END ##############################################
    # We have to start at the second row
    # 1st row is the text box for new todo items
    # 2nd row is the horizontal line
    row = 2

    # todo_list contains the list of items to be shown
    for index, todo in enumerate(todo_list):
        
        # a temporary placeholder to create a new sub-list
        # that contains 3 things that are shown with each iteration
        # 1. A StringVar that holds the text to be shown
        # 2. The Entry box (textbox)
        # 3. A checkmark shown as a button
        todo_list_entry = []

        # 1. StringVar 
        todo_list_entry.append(tk.StringVar())

        # 2. Textbox that sets the text to the stringvar above
        todo_list_entry.append(
            tk.Entry(window, textvariable=todo_list_entry[0]))

        # 2.1 Change the text to the actualy todo in the todo_list
        todo_list_entry[0].set(todo)
        
        # 2.2 
        todo_list_entry[1].config(state="disabled")

        # Set the textbox in the next row
        row = row + 1
        todo_list_entry[1].grid(row=row, column=0, padx=10, pady=3)
        
        # instead of using images, we will use unicode symbols like this
        # just search google for "unicode checkmark symbol" and you can 
        # just copy paste them (like any other text)
        text = ""
        todo_list_entry.append(tk.Button(window, text=text,
                                         command=lambda index=index: mark_complete_clicked(index)))

        # 3. Checkmark is also set in the same row adjacent to the text box
        todo_list_entry[2].grid(row=row, column=1, padx=10, pady=3)

        # All of these GUI widgets are appended to the array - todo_list_display
        todo_list_display.append(todo_list_entry)

render_list_items(todo_list)
window.mainloop()
```

In this section, we will add the buttons to make both of the following lists visible separately.
 - <kbd>open</kbd>
 - <kbd>completed</kbd>
 
To do this, we will create two buttons at the bottom that will show either of these lists. As soon as either of the buttons are clicked, the existing sets of todo widgets should be deleted and recreated depending on which button has been clicked. 

```python
import tkinter as tk
# from tkinter import ttk

todo_list = ["Lunch at 12:00 today", "Meeting preparation",
             "Third Item", "Fourth Item", "Fifth Item",
             "Sixth Item", "Seventh Item", "Eighth Item",
             "Ninth Item", "Tenth Item", "Eleventh Item",
             "Twelveth Item", "Thirteenth Item"] 

todo_list_complete = []

todo_list_display = []
####################### CHANGE START ###############################
open_button = None
completed_button = None
TODO_LIST_STATE = "OPEN"  # By default, we start the program with OPEN items
####################### CHANGE END #################################

# main window
window = tk.Tk()
window.title("TODO App")


todo_string = tk.StringVar()

todo_text = tk.Entry(window, textvariable = todo_string)

todo_text.grid(row=0, column = 0)

def addClicked () :
    ######################## CHANGE START ######################
    todo = todo_string.get()
    if todo == "":
        return
    todo_list.append(todo)
    render_list_items(todo_list)
    todo_string.set("")
    ######################## CHANGE END ########################

add_button = tk.Button(window, text="Add", command=addClicked)

add_button.grid(row=0, column = 1, padx=10, pady=10)

separator = tk.ttk.Separator(window)
separator.grid(row=1,columnspan=2, sticky="ew", pady=10)

def mark_complete_clicked(index):

    # add to todo_list_complete
    todo_list_complete.append(todo_list[index])

    # remove the clicked element from todo_list
    del todo_list[index]

    # re-render the list items
    render_list_items(todo_list)

####################### CHANGE START #######################
def mark_delete_clicked(index):
    del todo_list_complete[index]
    render_list_items(todo_list_complete)


def populate_open_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "OPEN"
    render_list_items(todo_list)


def populate_completed_items():
    global TODO_LIST_STATE
    TODO_LIST_STATE = "COMPLETED"
    render_list_items(todo_list_complete)    
    
####################### CHANGE END #######################    

def render_list_items(todo_list):

    for index, entry in enumerate(todo_list_display):
        entry[1].destroy()
        entry[2].destroy()

    todo_list_display.clear()
    
    ################## CHANGE START ###############################
    # by default variables are local in Python.
    global open_button
    global completed_button
    
    # destroy the buttons at the bottom
    if open_button != None:
        open_button.destroy()
    if completed_button != None:
        completed_button.destroy()    
    ################### CHANGE END ###############################
    row = 2

    # todo_list contains the list of items to be shown
    for index, todo in enumerate(todo_list):

        todo_list_entry = []

        # 1. StringVar 
        todo_list_entry.append(tk.StringVar())

        # 2. Textbox that sets the text to the stringvar above
        todo_list_entry.append(
            tk.Entry(window, textvariable=todo_list_entry[0]))

        # 2.1 Change the text to the actualy todo in the todo_list
        todo_list_entry[0].set(todo)
        
        # 2.2 
        todo_list_entry[1].config(state="disabled")

        # Set the textbox in the next row
        row = row + 1
        todo_list_entry[1].grid(row=row, column=0, padx=10, pady=3)
        
        ######################## CHANGE START ###############################
        text = ""
        if TODO_LIST_STATE == "OPEN":
            text = ""
            todo_list_entry.append(tk.Button(window, text=text,
                                             command=lambda index=index: mark_complete_clicked(index)))
        elif TODO_LIST_STATE == "COMPLETED":
            text = ""
            todo_list_entry.append(tk.Button(window, text=text,
                                             command=lambda index=index: mark_delete_clicked(index)))
        ######################### CHANGE END ###############################


        # 3. Checkmark is also set in the same row adjacent to the text box
        todo_list_entry[2].grid(row=row, column=1, padx=10, pady=3)

        # All of these GUI widgets are appended to the array - todo_list_display
        todo_list_display.append(todo_list_entry)
        
    ######################### CHANGE START ################################
    # Since we don't know the row numbers dynamically, add the buttons here only

    open_button = tk.Button(window, text="Open", 
                            command=populate_open_items)
    open_button.grid(row=row+1, columnspan=2,
                     sticky="W", padx=10, pady=(10, 10))

    completed_button = tk.Button(
        window, text="Completed", 
        command=populate_completed_items)
    completed_button.grid(row=row+1,  
                          columnspan=2, pady=(10, 10))
    ####################### CHANGE END ###################################

render_list_items(todo_list)
window.mainloop()
```
