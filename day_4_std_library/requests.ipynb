Say the browser is making a request to a server to get a list of blogs. There is a mock server at https://jsonplaceholder.typecode.com that can serve up dummy blogs for testers or learners like us.

<div class="text-center">
    
</div>

<br>

HTTP API works with in a <kbd>request</kbd> - <kbd>response</kbd> model. The client places a request with the server and the server comes back with a response. 

<br>

<div class="text-center">
    
</div>

<br>

Here is how we make the <kbd>GET</kbd> request with the requests library.

```python
import requests

posts = requests.get("https://jsonplaceholder.typicode.com/posts")
posts
```

<br>

<div class="text-center">
    
</div>

<br>

When you print the response object, it just prints the response code. In this case, <kbd>200</kbd> indicates a successful response code. But where is the actual response .. the list of blog posts ? It is available in the <kbd>content</kbd> of the response object.

```python
posts.content

## Sample output
##b'[\n  {\n    "userId": 1,\n    "id": 1,\n    "title": "sunt aut facere repellat provident occaecati excepturi optio reprehenderit",\n    "body": "quia et suscipit\\nsuscipit recusandae consequuntur expedita et cum\\nreprehenderit molestiae ut ut quas totam\\nnostrum rerum est autem sunt rem eveniet architecto"\n  },\n  {\n    "userId": 1,\n    "id": 2,\n    "title": "qui est esse",\n    "body": "est rerum tempore vitae\\nsequi sint nihil reprehenderit dolor beatae ea dolores neque\\nfugiat blanditiis voluptate porro vel nihil molestiae ut reiciendis\\nqui aperiam non debitis possimus qui neque nisi nulla"\n  },\n  
```

It produces a huge text in a json string format (actually, an array of json objects). Let's load it into python's JSON object.

```python
import json

posts_json =  json.loads(posts.content)
posts_json[0]
```

Ignore the language of the title and body. Just focus on the fact that we are getting the details of blog like the blog id, blog title, blog body etc. We know that the request is a success because the <kbd>status_code</kbd> returns a 200, which means OK. We will learn more about status codes in the upcoming section.

```python
posts.status_code
```

<div class="alert alert-secondary" id="Request-Methods">    
    <span class="topic"> &#10070; Request Methods</span>   
</div>

Previously we used only the <kbd>GET</kbd> method (also called <code>verb</code> ) to fetch resources from the server. GET can only "get" resources from the server. 

- What if we want to create resources ? Like create a blog post or create a user.
- Modify a blog post or comment associated with a blog post
- Delete a blog post or its comments. 

For these, we use the other methods available in standard HTTP APIs. Here are the most used ones

<table>
    <tr>
        <td> <b>HTTP Method</b> </td>
        <td> <b>Description</b> </td>
    </tr>
    <tr>
        <td> <kbd>GET</kbd> </td>
        <td> Used to "Get" resources from the HTTP API server. For example, get a list of blog posts. </td>
    </tr>
    <tr>
        <td> <kbd>POST</kbd> </td>
        <td> Used to "Create" new resources on the server. For example, create a new post using the data in the HTTP body. </td>
    </tr>
    <tr>
        <td> <kbd>PATCH</kbd> </td>
        <td> Used to "Update" an existing resource on the server. For example, change the post's subject. Since we refer to resources that are already existing in the server, we typically refer to the existing resource by some kind of an id. For example, to change an exist post, we refer to it's postId field. </td>
    </tr>
    <tr>
        <td> <kbd>DELETE</kbd> </td>
        <td> Used to "Delete" an existing resource on the server. For example, to remove a comment, we use the <b>delete</b> method by referring to the unique id of the comment to be deleted. </td>
    </tr>
    <tr>
        <td> <kbd>PUT</kbd> </td>
        <td> Used to "Replace" an existing resource on the server. For example, if you want to replace the entire blog post with another (different content), we use the <b>put</b> method by referring to the unique id of the post to be replaced. </td>
    </tr>
</table>

There are other less commonly used methods like <code>HEAD</code> , <code>OPTIONS</code> . We will not worry about them in this course. 

Let's see some examples of the most used methods described above. We have already seen the <kbd>GET</kbd> method in the section above. Let's now create a new post using the <kbd>POST</kbd> method. Don't confuse the <code>POST</code> method with the blog "post". They just happen to have the same name. But you can create a new User, a new Comment or any resource that the server allows using the <code>POST</code> method.

```python
res = requests.post('https://jsonplaceholder.typicode.com/posts', 
                    data = {"title": "Some Title 2",
                            "body" : "Some text in the body",
                            "userId" : 1})
```

There we go. We have just created a new post. Since this is a "mock" server, the post is not actually created. It is just "faked". Let's examine the status code.

```python
res.status_code
```

Succesful <code>GET</code> request returned a status code of <kbd>200</kbd>. However, a succesful <kbd>POST</kbd> request returned a status code of <kbd>201</kbd> - which means "successfully created". Now, let's update the title of an existing post. To refer to an existing post, we use the URL format 

https://jsonplaceholder.typicode.com/posts/{post_id}

where **post_id** refers to the id of the actual post in the server database. Since we want to change the **title** of the blog, we need to send the new title. In <kbd>PATCH</kbd> requests, we send data using the HTTP request <kbd>body</kbd>. Here is how we do it using the requests library."



```python
res = requests.patch('https://jsonplaceholder.typicode.com/posts/1', 
                    data = {"title": "Some Title 2"})
print (res.status_code)
print (json.loads(res.content))
```

To delete the entire post, use the <kbd>DELETE</kbd> method. It doesn't require any parameters in the body. 

```python
res = requests.delete('https://jsonplaceholder.typicode.com/posts/1')
print (res.status_code)
print (json.loads(res.content))
```

Great. We got a status code of 200, which means that the server able to delete the post with post id of 1. Once again, since this is a mock server, the actual data is deleted, but a correct code is returned if the request is properly made.

If you want the entire post to be updated with a new post content, use the <kbd>PUT</kbd> request.

```python
data = {"title" : "New Title","body" : "New blog body" , "userId" : 1 }
res = requests.put('https://jsonplaceholder.typicode.com/posts/1')
print (res.status_code)
print (json.loads(res.content))
```

<div class="alert alert-secondary" id="Status-Codes">    
    <span class="topic"> &#10070; Status Codes</span>   
</div>

Every HTTP request gets a response and each response is coded. In general, the response codes can be categorized as follows

<table>
    <tr>
        <td> <b>HTTP Status Code</b></td>
        <td> <b>Meaning</b></td>
        <td> <b>More info</b> </td>
    </tr>
    <tr>
        <td> 2xx </td>
        <td> Success </td>
        <td> These codes generally mean that the HTTP request has succeeded. For example, a <kbd>201</kbd> means that a requested resource has been created.
    </tr>
    <tr>
        <td> 3xx </td>
        <td> Redirection </td>
        <td> Sometimes, resources are not available at the requested URL. Instead they are served from another location. For example, <kbd>301</kbd> indicates that the requested resource has been permanently moved.
    </tr>
    <tr>
        <td> 4xx </td>
        <td> Client Errors </td>
        <td> The client could make errors when making a request. For example, a <kbd> 401 </kbd> implies that the user is asking for a resources without the right authorization. Or a <kbd>404</kbd> means that the requested resoure does not exist.
            
            
    </tr>
    <tr>
        <td> 5xx </td>
        <td> Server Errors </td>
        <td> Sometimes, the server might crash serving a request. For example, if the server doesn't know how to handle a request, it might throw a <kbd>501</kbd> error.
    </tr>
    </table>

Let's see some examples below. 

For example, we know that there is no blog with id 200 in the sample API above. So, when we ask for it, the server knows that we are asking for a wrong blog. So, it throws a <kbd>404</kbd> error. 

```python
req = requests.get("https://jsonplaceholder.typicode.com/posts/200")
req.status_code
```

However, if you request a blog post (resource) that is available and can be served without authorization, it sends a status code of <kbd>200</kbd> which means <kbd>ok</kbd>.

```python
req = requests.get("https://jsonplaceholder.typicode.com/posts/1")
req.status_code
```

<div class="alert alert-secondary" id="Query-Parameters">    
    <span class="topic"> &#10070; Query Parameters</span>   
</div>

Sometimes, <kbd>GET</kbd> requests can be further filtered if the server allows it. For example, We know how to get all the posts or users and also how to get a specific post content and a specific user details. 

```python
req = requests.get("https://jsonplaceholder.typicode.com/users")
users_json = json.loads(req.content)
users_json[0]
```

```python
req = requests.get("https://jsonplaceholder.typicode.com/posts")
posts_json = json.loads(req.content)
posts_json[0]
```

What if we want to get all the posts that belong to a particular user ? Most HTTP APIs implement them as URL parameters. 

```python
req = requests.get("https://jsonplaceholder.typicode.com/posts?userId=2")
posts_users_json = json.loads(req.content)
posts_users_json[0]
```

A more elegant way of doing this would be as follows. 

```python
params = {"userId" : 2}
req = requests.get("https://jsonplaceholder.typicode.com/posts", params = params)
posts_users_json = json.loads(req.content)
posts_users_json[0]
```

We would get the exact same response. 

<code>Note</code> - Query parameters are typically only sent for <kbd>GET</kbd> requests. Other type of requests (like <kbd>POST</kbd>, <kbd>PUT</kbd> etc) that send data over to the server do so via the request body.

<div class="alert alert-secondary">
    <h4 class="alert-heading">&#9998; &nbsp; Learning</h4>  
    <hr>
<ul class="list-group-flush list-group-numbered">
  <li class="list-group-item"><kbd >requests</kbd> library can be used to interact with servers that expore HTTP APIs.</li>
    <li class="list-group-item">Each HTTP request has a <code>method</code> that specifies to the server what kind of action needs to be taken. Examples are <kbd>GET</kbd> , <kbd>POST</kbd>, <kbd>PUT</kbd>, <kbd>PATCH</kbd>, <kbd>DELETE</kbd> etc. The requests class has methods for each of these methods.</li>
  <li class="list-group-item">Every HTTP request from the client gets a response from the server. Every response data includes a <kbd>response code</kbd> and data if any. <kbd>content</kbd> attribute of the <kbd>response</kbd> object contains the actual content sent over from the server. </li>
  <li class="list-group-item">Response codes are numbers like 200, 401, 404 etc. Each of these numbers have a specific meaning. <kbd>status_code</kbd> property of the response object contains the response code. </li>
    <li class="list-group-item"><kbd>query parameters</kbd> can be used in the <kbd>GET</kbd> method to provide additional info or provide additional constraints to the requested data.</li>
   </ul>
